```
import random
import math
```

```
inside_circle = 0
t_darts = 10000 

for i in range(0, t_darts):
  x = random.random()**2 # ** is used to calculate powers
  y = random.random()**2
  if math.sqrt(x + y) <= 1.0:
    inside_circle += 1


pi = (float(inside_circle) / t_darts) * 4

print(pi)
```

```
import numpy as np #https://www.science-emergence.com/Articles/How-to-plot-a-circle-in-python-using-matplotlib-/
import matplotlib.pyplot as plt

theta = np.linspace(0, 2*np.pi, 100)

r = np.sqrt(1.0)

x1 = r*np.cos(theta)
x2 = r*np.sin(theta)

fig, ax = plt.subplots(1)

ax.plot(x1, x2)
ax.set_aspect(1)

plt.xlim(-1.25,1.25)
plt.ylim(-1.25,1.25)

plt.grid(linestyle='--')

plt.title('How to plot a circle with matplotlib ?', fontsize=8)

plt.savefig("plot_circle_matplotlib_01.png", bbox_inches='tight')

plt.show()
```

```
# Import Packages
import numpy as np
import time
import matplotlib.pyplot as plt
```

```
#define a function to approximate the area of a cirlce using a Monte Carlo simulation
def simulation(n):
  darts_in_circle = 0
  area_of_square = 4

  # Generate random array of coordinates
  darts_thrown = np.random.rand(int(n),2)

  # Check to see if it is in the radius of the circle using the distance formula
  for dart in darts_thrown:
    if dart[0]**2+ dart[1]**2 < 1:

      #if it is in the circle, add it to a counter
      darts_in_circle += 1

  # divide total number in the circle by the total number in the random array and multiply by the area of the square
  approx_area_of_circle = (darts_in_circle / len(darts_thrown))*area_of_square  
  #print(approx_area_of_circle)
  return(approx_area_of_circle)
```

```
#create an array of different numbers darts thrown to test processing time
number_of_darts_thrown = np.linspace(1,1000000,20, dtype=int);
processing_time = []

#go through list of different darts thrown
for dart in number_of_darts_thrown:
  #for each one, start a time
  start_time = time.time()
  #compute the approximation
  simulation(dart)
  #determine how long each simulation takes and append it to an array
  processing_time.append(time.time() - start_time)

print(processing_time)
```

```
#plot processing time (y) with respect to number of darts thrown (x)
plt.plot(number_of_darts_thrown, processing_time, label="Line Graph")
plt.scatter(number_of_darts_thrown, processing_time, color="red", label = "Data Points")

#can alternatively be done with plt.plot(number_of_darts_thrown, processing_time, '.', number_of_darts_thrown, processing_time)

#comparing it with a regression
z = np.polyfit(number_of_darts_thrown,processing_time, 4)
p = np.poly1d(z)
xp = np.linspace(0, 1000000, 10000)
plt.plot(xp, p(xp), color="orange", label="Polynomial Regression")

#plot additional features on graph
plt.legend()
plt.ylabel('Time taken (s) for Monte Carlo simulation')
plt.xlabel('Number of Darts Thrown ($10^6$)')
plt.title('Number of Darts ($10^6$) vs. Time (s)')
```

```
import random
import numpy as np
import time
import matplotlib.pyplot as plt

#define a function to approximate the area of a cirlce using a Monte Carlo simulation
def simulation(n):
  darts_in_circle = 0
  area_of_square = 4

  # Generate random array of coordinates
  darts_thrown = np.random.rand(int(n),2)

  # Check to see if it is in the radius of the circle using the distance formula
  for dart in darts_thrown:
    if dart[0]**2+ dart[1]**2 < 1:

      #if it is in the circle, add it to a counter
      darts_in_circle += 1

  # divide total number in the circle by the total number in the random array and multiply by the area of the square
  approx_area_of_circle = (darts_in_circle / len(darts_thrown))*area_of_square  
  #print(approx_area_of_circle)

number_of_darts_thrown = np.linspace(1,1000000,20, dtype=int);
processing_time = []

#go through list of different darts thrown
for dart in number_of_darts_thrown:
  #for each one, start a time
  start_time = time.time()
  #compute the approximation
  simulation(dart)
  #determine how long each simulation takes and append it to an array
  processing_time.append(time.time() - start_time)

print(processing_time)

#plot processing time (y) with respect to number of darts thrown (x)
plt.plot(number_of_darts_thrown, processing_time, label="Line Graph")
plt.scatter(number_of_darts_thrown, processing_time, color="red", label = "Data Points")

#can alternatively be done with plt.plot(number_of_darts_thrown, processing_time, '.', number_of_darts_thrown, processing_time)

#comparing it with a regression
z = np.polyfit(number_of_darts_thrown,processing_time, 4)
p = np.poly1d(z)
xp = np.linspace(0, 1000000, 10000)
plt.plot(xp, p(xp), color="orange", label="Polynomial Regression")

#plot additional features on graph
plt.legend()
plt.ylabel('Time taken (s) for Monte Carlo simulation')
plt.xlabel('Number of Darts Thrown ($10^6$)')
plt.title('Number of Darts ($10^6$) vs. Time (s)')
```
